# 2. Data Quality check and Handling Missing values

```python
# Checking for Categorical and continuous variables in application_data

app_data.nunique().sort_values()
```

```python
# Finding the percentage of missisng values for all the columns.

percent_missing = app_data.isnull().sum() * 100 / len(app_data)
```

```python
# Understandig the percentage variation

percent_missing.describe()
```

```python
# Ignoring the columns with high missing percentages and selecting the columns which has lower percentage.
# From the above cell we can infer that the least percentage which is greater than zero is 13.501631.
# Hence select all the columns which have missing value percentage less than the mean and assign it to 'new' dataframe.

new = percent_missing[percent_missing <= percent_missing.mean()].keys()
```

```python
# Creating a new dataframe app_data that stores the variable with missing percentage less that the mean.
app_data = app_data[new]
```

```python
# selecting 5 columns from app_data.

app_data[['AMT_ANNUITY','AMT_GOODS_PRICE','NAME_TYPE_SUITE','EXT_SOURCE_2','CNT_FAM_MEMBERS']]
```

## Imputation of missing values for the above columns
- AMT_ANNUITY: Best metric to impute the missing values is by the MEAN of the column 
- AMT_GOODS_PRICE: Best metric to impute the missing values is by the MEAN of the column 
- NAME_TYPE_SUITE: Best metric to impute the missing values is by the MODE of the column 
- EXT_SOURCE_2: Best metric to impute the missing values is by the MEAN of the column
- CNT_FAM_MEMBERS:  Best metric to impute the missing values is by the MODE of the column

```python
#checking datatypes of the columns for app_data.

app_data.info()
```

## Columns for which the datatype needs to be changed.
- CNT_FAM_MEMBERS: Float to integer datatype
- DAYS_BIRTH: Client's age from negative to positive and converting in terms of year.
- DAYS_EMPLOYED: Changing from negative to positive and converting to years
- DAYS_REGISTRATION: Changing from negative to positive and converting to years
- DAYS_ID_PUBLISH: Changing from negative to positive and converting to years 


```python
# Changing the datatype of CNT_FAM_MEMBERS from float to int. 

app_data['CNT_FAM_MEMBERS'] = app_data['CNT_FAM_MEMBERS'].round().astype('Int64')
```

```python
# Converting DAYS_BIRTH into year format by dividing with 356 and using abs() function.

app_data['DAYS_BIRTH'] = app_data['DAYS_BIRTH'].abs()/365
app_data['DAYS_BIRTH'] = round(app_data['DAYS_BIRTH'])
app_data['DAYS_BIRTH'] = app_data['DAYS_BIRTH'].astype('Int64')
```

```python
# Converting DAYS_EMPLOYED into year format by dividing with 356 and using abs() function.

app_data['DAYS_EMPLOYED'] = app_data['DAYS_EMPLOYED'].abs()/365
app_data['DAYS_EMPLOYED'] = round(app_data['DAYS_EMPLOYED'],1)
```

```python
# Converting DAYS_REGISTRATION into year format by dividing with 356 and using abs() function.

app_data['DAYS_REGISTRATION'] = app_data['DAYS_REGISTRATION'].abs()/365
app_data['DAYS_REGISTRATION'] = round(app_data['DAYS_REGISTRATION'] , 1)
```

```python
# Converting DAYS_ID_PUBLISH into year format by dividing with 356 and using abs() function.

app_data['DAYS_ID_PUBLISH'] = app_data['DAYS_ID_PUBLISH'].abs()/365
app_data['DAYS_ID_PUBLISH'] = round(app_data['DAYS_ID_PUBLISH'],1)
```

# Correlation

```python
#Correlation for target_0

corr_0 = target_0.notnull().corr()

corr_0.where(np.triu(np.ones(corr_0.shape), k=1).astype(np.bool))
corr_0 = corr_0.unstack().reset_index()
corr_0.columns = ['Var1', 'Var2', 'Correlation']
corr_0.dropna(subset=['Correlation'], inplace=True)
corr_0['Correlation_abs'] = corr_0['Correlation'].abs()
corr_0['Correlation'] = corr_0['Correlation'].abs()
corr_0.sort_values("Correlation_abs", ascending=False, inplace=True)
corr_0 = corr_0[corr_0['Var1'] != corr_0['Var2']]
corr_0 = corr_0.iloc[::2]
corr_0.head(10)
```

```python
#Correlation for target 1

corr_1 = target_1.notnull().corr()

corr_1.where(np.triu(np.ones(corr_1.shape), k=1).astype(np.bool))
corr_1 = corr_1.unstack().reset_index()
corr_1.columns = ['Var1', 'Var2', 'Correlation']
corr_1.dropna(subset=['Correlation'], inplace=True)
corr_1['Correlation_abs'] = corr_1['Correlation'].abs()
corr_1['Correlation'] = corr_1['Correlation'].abs()
corr_1.sort_values("Correlation_abs", ascending=False, inplace=True)
corr_1 = corr_1[corr_1['Var1'] != corr_1['Var2']]
corr_1 = corr_1.iloc[::2]
corr_1.head()
```

# Results of Correlation

-  We can also see that AMT_GOODS_PRICE variable is correlated with CNT_FAM_MEMBERS in target_0

### AMT_CREDIT_CHANGE : The difference between the amount of loan applied and got sanctioned. Influence of the other variable’s like income, exit score(CIBIL score) and attitude of the client towards previous loan can make the value negative and positive otherwise.


```python
prev_data['AMT_CREDIT_CHANGE'] =  prev_data['AMT_CREDIT'] - prev_data['AMT_APPLICATION'] 
```

- Cash loans had always increased in AMT_CREDIT with respect to the AMT_APPLLICATION
- There is a major inccrease in the AMT_CREDIT for the revolving loans
